# Data Visualization

Import required python package.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Optional if you use seaborn
plt.style.use('seaborn')
```

## Load Data

The dataset can be downloaded from here: https://archive.ics.uci.edu/ml/datasets/iris.

```python
df_iris = pd.read_csv('./datasets/iris.data')
df_iris
```

Define the column name, so the first row is not converted into header. (it's necessary if you load from .data file, it should be okay if you load from .csv file). You can obtain datasets information & attribute information from above link where you can download iris dataset.

```python
df_iris = pd.read_csv('./datasets/iris.data', names=['sepal_length', 'sepal_width',
                                                    'petal_length', 'petal_width', 'class'])
df_iris
```

## Data Visualization

### Histogram - Distribution

Create Histogram to check distribution of each class in iris dataset.

```python
df_iris['class'].hist()
plt.show()
```

Create scatter plot to check distribution & relation between two variables.

```python
df_iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
df_iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

```python
df_iris
```

## Distplot

You could also use distplot to check the distribution.

```python
sns.distplot(df_iris['sepal_length'])
plt.show()
```

## Boxplot

Use boxplot to check distribution & comparison.

```python
sns.boxplot(data=df_iris)
plt.xlabel('Features')
plt.ylabel('Length in cm')
plt.show()
```

Comparison of petal_length in each class of iris dataset, try another variable like sepal_width, petal_length, etc.

```python
sns.boxplot(data=df_iris, x='class', y='sepal_length')
plt.xlabel('Class')
plt.ylabel('Length in cm')
plt.show()
```

## Pair Plot

To check relation, distribution, and comparison in one go (please be in mind, it depends on how many variable that you have, it could take some time to generate pair plot if you have a lot variable in your dataset).

```python
sns.pairplot(data=df_iris)
plt.show()
```

You can also group data based on each class in a separate color.

```python
sns.pairplot(data=df_iris, hue='class')
plt.show()
```

## Heatmap

To find correlation/relation between variables.

```python
sns.heatmap(data=df_iris.corr(), annot=True)
plt.show()
```
