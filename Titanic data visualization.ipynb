# Titanic Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
df_titanic = pd.read_csv('titanic.csv')
df_titanic['Age'] = df_titanic['Age'].astype('int64')
df_titanic.head(10)
```

```python
df_titanic.describe()
```

Distribusi dari Umur Penumpang Titanic

```python
sns.distplot(df_titanic['Age'])
plt.show()
```

Distribusi Banyak penumpang berdasarkan jenis kelamin

```python
df_titanic['Sex'].hist()
plt.show()
```

Distribusi Banyak Penumpang berdasarkan Kelas Penumpang

```python
sns.catplot(x='Pclass',kind='count',data=df_titanic)
```

Perbandingan banyak penumpang berdasarkan kelas penumpang dan jenis kelamin

```python
sns.catplot(x='Pclass',kind='count',hue='Sex',data=df_titanic)
```

Survival by Sex

```python
sns.catplot(x='Survived',kind='count',hue='Sex',data=df_titanic)
```

Survival by Age

```python
sns.boxplot(data=df_titanic, x='Survived', y='Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

Survival by Fare

```python
sns.boxplot(data=df_titanic, x='Survived', y='Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

Survival by Passenger Class

```python
sns.catplot(x='Pclass',kind='count',hue='Survived',data=df_titanic)
```

Pairplot

```python
sns.pairplot(data=df_titanic, hue='Survived')
plt.show()
```

Heatmap

```python
sns.heatmap(data=df_titanic.corr(), annot=True)
plt.show()
```
